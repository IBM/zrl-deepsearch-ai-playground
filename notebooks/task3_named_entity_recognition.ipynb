# Import libraries

```python
import pandas as pd
import spacy
from spacy import displacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")
from pprint import pprint
```

# Let's try

```python
# text from some bbc news article
text = """Survivors of Afghanistan's deadliest earthquake in two decades say they have nothing to eat, no shelter, and fear a possible cholera outbreak. The BBC's Secunder Kermani reports from Paktika province, the hardest hit by the disaster."""
doc = nlp(text)
```

```python
pprint([(X.text, X.label_) for X in doc.ents])
```

```python
spacy.displacy.render(doc, style="ent",jupyter=True)
```

```python
spacy.explain('NNS')
```

# Let's look at the analysis a bit more deeply

```python
def tabulate_result(doc):
    df_text = pd.DataFrame()
    columns = ['text', 'ent', 'pos','tag','ent_explain', 'pos_explain', 'tag_explain']
    df_text.reindex(columns=columns)
    for index in range(len(doc)):
        token = doc[index]
        df_text.loc[index, 'text'] = token.text
        df_text.loc[index, 'ent'] = token.ent_type_        
        df_text.loc[index, 'pos'] = token.pos_
        df_text.loc[index, 'tag'] = token.tag_
        if token.ent_type_ == '':
                df_text.loc[index, 'ent_explain'] = ''
        else:
                df_text.loc[index, 'ent_explain'] = spacy.explain(token.ent_type_)
        df_text.loc[index, 'pos_explain'] = spacy.explain(token.pos_)
        df_text.loc[index, 'tag_explain'] = spacy.explain(token.tag_)
#         df_text.loc[index, 'dependency'] = token.dep_
    return df_text
```

```python
tabulate_result(doc)
```
